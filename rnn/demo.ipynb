```python
df = pd.read_excel("toy_data_easy.xlsx")
df.drop("sample_id",axis=1,inplace=True)
print(df.shape)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.drop("target",axis=1)[:100]
y = df["target"][:100]

scaler = StandardScaler()
scaler.fit_transform(X)


X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
X_train.shape
```

```python
model = RNN(10,[16,1])
model
```

```python
epochs = 100
lr = 0.001
alpha = 0
loss_fn = MSELoss()
hidden_state = [Value(0) for _ in range(10)]



for epoch in range(epochs):
  model.hidden_state = [Value(0.0) for _ in range(model.nin)]
  input = [list(map(Value,i)) for i in X_train.values]
  y_pred = [model(val) for val in input]



  data_loss = loss_fn(y_pred,y_train)
  regu_loss = alpha * sum([pm * pm for pm in model.parameters()])
  loss = data_loss + regu_loss

  max_grad = 1.0
  for pm in model.parameters():
    pm.grad = max(-max_grad, min(max_grad, pm.grad))
  loss.backward()
  for pm in model.parameters():
    pm.data -= lr * pm.grad

  if(epoch % 10 == 0):
    print(
        f"for epoch {epoch} : train loss {loss}"
    )
```
